```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"F:\I-Neuron\22-OCT-END-TO-END-PROJECT\notebooks\data\gemstone.csv")
```

```python
data.drop(columns="id")
```

```python
# dependent ana independent column
x = data.drop(columns="price")
y=data[["price"]]
```

```python
# segregate categorical and numerical column
cat_column=x.columns[x.dtypes=="object"]
num_column=x.columns[x.dtypes!="object"]
x_numercal_data=x[num_column]
x_categorical_data=x[cat_column]
x_categorical_data
```

```python
x_cat_data=x.select_dtypes(include="object")
x_num_data=x.select_dtypes(exclude="object")
```

```python
x_cat_data_col=x_cat_data.columns
x_num_data_col=x_num_data.columns

x_num_data_col


```

```python
cut_categorical = ["Fair","Good","Very Good","Premium","Ideal"]
clarity_categorical = ["I1","SI2","SI1","VS2","VS1","VVS2","VVS1","IF"]
color_categorical = ["D","E","F","G","H","I","J"]
```

```python
from sklearn.impute import SimpleImputer  #handling missing value
from sklearn.preprocessing import OrdinalEncoder  #ordinal encoder
from sklearn.preprocessing import StandardScaler # handling feature scaling
#preprocessing
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
numeric_pipeline=Pipeline(


    steps=[

        ("imputer",SimpleImputer()),
        ("scaler",StandardScaler())





    ]
)
```

```python
categorical_pipeline=Pipeline(




            steps=[

        ("imputer",SimpleImputer(strategy='most_frequent')),
        ("ordinalencoder",OrdinalEncoder(categories=[cut_categorical,color_categorical,clarity_categorical]))



    ]

)
```

```python
preprocessor=ColumnTransformer(


[

('num_pipeline',numeric_pipeline,x_num_data_col),
('cat_pipeline', categorical_pipeline,x_cat_data_col)

]
)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
preprocessor.fit_transform(x_train)
```

```python
preprocessor.transform(x_test)
```

```python
preprocessor.get_feature_names_out()
```

```python
x_train=pd.DataFrame(preprocessor.fit_transform(x_train),columns=preprocessor.get_feature_names_out())
```

```python
x_test=pd.DataFrame(preprocessor.transform(x_test),columns=preprocessor.get_feature_names_out())
```

```python
# linear regression
# ridge regression
# lasso regression
# elastic regression
from sklearn.linear_model import LinearRegression,Lasso,ElasticNet,Ridge
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models={
    'linearReression': LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elastic':ElasticNet()}
```

```python
trained_model_list=[]
model_list=[]
r2_list=[]
```

```python
import numpy as np
def evaluate_model(true,predict):
    mae=mean_absolute_error(true,predict)
    mse=mean_squared_error(true,predict)
    rmse=np.sqrt(mean_squared_error(true,predict))
    r2_square=r2_score(true,predict)
    return mae,rmse,r2_square
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    #prediction
    y_predict=model.predict(x_test)

    #this is validation test score

    mae,rmse,r2_square=evaluate_model(y_test,y_predict)

    print(list(models.keys())[i])
    model_list.append( (list(models.keys())[i]))

    print ("model training performance")
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2_SCORE:",r2_square)
    r2_list.append(r2_square)

    print("="*35)
    print("\n")
```
